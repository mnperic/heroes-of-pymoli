```python
# Import dependencies
import pandas as pd
import numpy as np

# File to Load
file_to_load = "/Users/minoperic/Data Analytics/pandas-challenge-clone/HeroesOfPymoli/purchase_data.csv"

# Store .csv into pandas dataframe
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items
number_unique_items = len(purchase_data["Item ID"].value_counts())

# Average price
average_price = (purchase_data["Price"].mean())

# Number of purchases
number_purchases = (purchase_data["Price"].count())

# Total revenue
total_revenue = (purchase_data["Price"].sum())

# Display summary dataframe
purchase_overview = {"Number of Unique Items": [number_unique_items], "Average Price":[average_price], "Number of Purchases":[number_purchases], "Total Revenue":[total_revenue]}
purchase_overview_df = pd.DataFrame(purchase_overview)

# Format dataframe columns (2 decimal places)
purchase_overview_df["Average Price"] = purchase_overview_df["Average Price"].map("${:.2f}".format)
purchase_overview_df["Total Revenue"] = purchase_overview_df["Total Revenue"].map("${:.2f}".format)

purchase_overview_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_data = purchase_data.groupby(["Gender"])
gender_data.count()

# Total count by gender
male = len(purchase_data.loc[purchase_data["Gender"]=="Male", :])
female = len(purchase_data.loc[purchase_data["Gender"]=="Female", :])
other = len(purchase_data.loc[purchase_data["Gender"]=="Other", :])


# Percentage of players by gender
percent_male = (male/(male + female + other)) * 100
percent_female = (female/(male + female + other)) * 100
percent_other = (other/(male + female + other)) * 100


# Display summary dataframe
gender_data_df = pd.DataFrame(
[[percent_male, male],
[percent_female, female],
[percent_other, other]],
index=["Male","Female", "Other"],
columns=["Percentage of Players", "Total Count"])

# Format dataframe columns (2 decimal places)
gender_data_df["Percentage of Players"] = gender_data_df["Percentage of Players"].map("{:.2f}%".format)
gender_data_df
```


## Purchasing Analysis (Gender)

* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results

* Display a preview of the summary data frame



```python
total_purchase = purchase_data.groupby("SN")["Price"].sum()
purchase_count = purchase_data.groupby("SN")["Price"].count()
purchase_avg = purchase_data.groupby("SN")["Price"].mean()

# Create summary Data frame
spender_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": purchase_avg, "Total Purchase Value": total_purchase})

# Sort by Total Purchase Value
spender_df.sort_values("Total Purchase Value", ascending = False, inplace=True)

# Format columns (2 decimal places)
spender_df["Average Purchase Price"] = spender_df["Average Purchase Price"].map("${:.2f}".format)
spender_df["Total Purchase Value"] = spender_df["Total Purchase Value"].map("${:.2f}".format)

spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Calculate most popular item
popular_count = purchase_data.groupby("Item Name")["Price"].count()
popular_price = purchase_data.groupby("Item Name")["Price"].mean()
popular_total = purchase_data.groupby("Item Name")["Price"].sum()

# Create summary DataFrame
popular_df = pd.DataFrame({"Purchase Count": popular_count, "Average Item Price": popular_price, "Total Purchase Value": popular_total})

# Sort by Purchase Count 
popular_df.sort_values("Purchase Count", ascending = False, inplace=True)

# Format columns (2 decimal places)
popular_df["Average Item Price"] = popular_df["Average Item Price"].map("${:.2f}".format)
popular_df["Total Purchase Value"] = popular_df["Total Purchase Value"].map("${:.2f}".format)

popular_df.head()
```
